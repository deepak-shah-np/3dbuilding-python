```python
import geopandas as gpd
DATA_URL = "https://data.sfgov.org/resource/ynuv-fyni.geojson"
gdf = gpd.read_file(DATA_URL)
```

```python
gdf.head()
```

```python
select_one = gdf[gdf.area_id.eq('598')]
```

```python
select_one
```

```python
"""
GeoJsonLayer
===========

Property values in Vancouver, Canada, adapted from the deck.gl example pages. Input data is in a GeoJSON format.
"""


import pydeck as pdk




INITIAL_VIEW_STATE = pdk.ViewState(latitude=37.7155422, longitude=-122.4062674, zoom=16, max_zoom=20, pitch=45, bearing=0)


geojson = pdk.Layer(
    "GeoJsonLayer",
    select_one,
    opacity=0.9,
    stroked=False,
    filled=True,
    extruded=True,
    wireframe=True,
    get_elevation='10',
    get_fill_color="[128,128,128]",
    get_line_color=[255, 255, 255],
    coverage=1,
    pickable=True
)

add_polygon = pdk.Layer(
    "GeoJsonLayer",
    select_one,
    opacity=0.4,
    stroked=False,
    filled=True,
    extruded=True,
    wireframe=True,
    get_elevation='20',
    get_fill_color="[255, 255, 0*0.3]",
    get_line_color=[255, 255, 255],
    auto_highlight=True,

     coverage=1,
)
pdk.Layer


r = pdk.Deck(layers=[ geojson], initial_view_state=INITIAL_VIEW_STATE)
r1 = pdk.Deck(layers=[ geojson,add_polygon], initial_view_state=INITIAL_VIEW_STATE)

r.to_html("geojson_layer.html")
# r1.to_html("geojson_layer1.html")
```
